# Isaac Robot Loader
Dieses Notebook ist ein Startpunkt, um eine vorhandene USD-Datei eines Roboters in Isaac Sim zu laden.
Laut Isaac-Lab-Dokumentation ist der empfohlene Weg für neue Roboter meist: URDF in USD konvertieren und danach die USD-Datei referenzieren. [web:191][web:65]

```python
ROBOT_USD = '/absolute/path/to/your/robot.usd'
HEADLESS = True
ROBOT_PRIM_PATH = '/World/Robot'
PHYSICS_DT = 1/200
RENDER_DT = 1/60
```

```python
from pathlib import Path
robot_usd = Path(ROBOT_USD).expanduser().resolve()
assert robot_usd.exists(), f'USD nicht gefunden: {robot_usd}'
```

```python
from isaacsim import SimulationApp
simulation_app = SimulationApp({'headless': HEADLESS})
```

```python
from omni.isaac.core import World
from omni.isaac.core.utils.stage import add_reference_to_stage
from omni.isaac.core.utils.prims import get_prim_at_path
from omni.isaac.core.articulations import Articulation
```

```python
world = World(stage_units_in_meters=1.0, physics_dt=PHYSICS_DT, rendering_dt=RENDER_DT)
world.scene.add_default_ground_plane()
add_reference_to_stage(usd_path=str(robot_usd), prim_path=ROBOT_PRIM_PATH)
prim = get_prim_at_path(ROBOT_PRIM_PATH)
assert prim.IsValid(), f'Robot-Prim konnte nicht geladen werden: {ROBOT_PRIM_PATH}'
robot = Articulation(prim_path=ROBOT_PRIM_PATH, name='robot')
world.scene.add(robot)
world.reset()
print('DOFs:', robot.num_dof)
```

```python
for _ in range(300):
    world.step(render=not HEADLESS)
```

```python
simulation_app.close()
```
