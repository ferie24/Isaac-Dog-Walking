# Unitree A1 in Isaac Sim
Dieses Notebook laedt den eingebauten Unitree-A1-Asset aus Isaac Sim. Die Robot-Assets-Dokumentation nennt dafuer den Pfad `Robots/Unitree/A1/a1.usd`, und die Quadruped-Extension nennt A1 als unterstuetzten 12-DOF-Quadruped. [web:213][web:142]

```python
HEADLESS = True
ROBOT_PRIM_PATH = '/World/A1'
PHYSICS_DT = 1/200
RENDER_DT = 1/60
```

```python
from isaacsim import SimulationApp
simulation_app = SimulationApp({'headless': HEADLESS})
```

```python
from isaacsim.storage.native import get_assets_root_path
from omni.isaac.core import World
from omni.isaac.core.utils.stage import add_reference_to_stage
from omni.isaac.core.utils.prims import get_prim_at_path
from omni.isaac.core.articulations import Articulation
```

```python
assets_root = get_assets_root_path()
assert assets_root is not None, 'Isaac-Sim-Assets konnten nicht gefunden werden.'
A1_USD = assets_root + '/Isaac/Robots/Unitree/A1/a1.usd'
A1_USD
```

```python
world = World(stage_units_in_meters=1.0, physics_dt=PHYSICS_DT, rendering_dt=RENDER_DT)
world.scene.add_default_ground_plane()
add_reference_to_stage(usd_path=A1_USD, prim_path=ROBOT_PRIM_PATH)
prim = get_prim_at_path(ROBOT_PRIM_PATH)
assert prim.IsValid(), f'A1-Prim konnte nicht geladen werden: {ROBOT_PRIM_PATH}'
robot = Articulation(prim_path=ROBOT_PRIM_PATH, name='unitree_a1')
world.scene.add(robot)
world.reset()
print('DOFs:', robot.num_dof)
```

```python
for _ in range(300):
    world.step(render=not HEADLESS)
```

```python
simulation_app.close()
```
